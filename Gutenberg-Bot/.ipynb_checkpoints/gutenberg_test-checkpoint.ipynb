```python
import xml.etree.ElementTree as ET  
import os
from os import path
import rdflib
import bs4
import pandas as pd

def get_all_files():

    search_path = os.listdir("Data_Machine_readable")
    xml_list = []

    #Check if file containing directories for all xml files exists -> if not searches folder 
    if  path.exists("file.txt"):
        with open("file.txt", "r") as f:
            print("File Found")
            for line in f:
                xml_list.append(line.strip())

    else:    

        for dirpath, dirnames, filenames in os.walk("Data_Machine_readable"):
            if filenames:
                print(filenames[0],type(filenames))
                xml_list.append(os.path.join(dirpath,filenames[0]))
            
        with open("file.txt", "w") as f:
            for s in xml_list:
                f.write(str(s) +"\n")
    
    return xml_list


# Start "main-loop"

links = get_all_files()
Book_Archive = []

df1 = pd.read_pickle('123.pkl')

for i,n in enumerate(links[1:30]):

    tree = ET.parse(n)
    root = tree.getroot()
    print(i)
    panda_ser 
    
    my_dict = {}
    pandas_series = []
    
    for e in tree.findall('.//'):
        
        tags  = e.tag.split("}")[-1]
        atr  = e.attrib
        text = e.text
        
         
        panda_series.append([tags,text])
        
        
    dict_book = panda_series
    
    for n in dict_book:
        my_dict.update({n[0]:n[1]})
        
        
    #print(type(my_dict),my_dict)
    my_dict
    Book_Archive.append(my_dict)
    
    #panda_series = pd.Series(data= my_dict.items(),index = my_dict.keys())
    #Book_Archive.append(panda_series)
    #print(type(Book_Archive[0]),Book_Archive[0])
    
test_data = pd.DataFrame(Book_Archive)     
print(test_data.columns)
print(test_data.alias,test_data.title)
test_data.to_pickle('123.pkl') 
```
